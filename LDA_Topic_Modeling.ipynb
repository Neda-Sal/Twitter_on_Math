## Try LDA for topic modeling

```python
import pandas as pd
import numpy as np
from sklearn.feature_extraction.text import TfidfVectorizer, ENGLISH_STOP_WORDS
from sklearn.decomposition import LatentDirichletAllocation

# logging (set to INFO)
import logging
logging.basicConfig(format='%(asctime)s : %(levelname)s : %(message)s', level=logging.INFO)

import pyLDAvis
import pyLDAvis.sklearn
pyLDAvis.enable_notebook()
```

```python
tweets = pd.read_pickle('math_tweets_no_duplicates.pkl')
tweets.head()
```

```python
corpus = tweets['clean_english']
```

```python
stop_words = ENGLISH_STOP_WORDS.union(['math', 'rt', 'https'])
```

```python
tfidf = TfidfVectorizer(stop_words=stop_words)
tweet_word_matrix = tfidf.fit_transform(corpus)
#vocab = tfidf.get_feature_names()
```

```python
tweet_word_matrix.shape
```

```python
lda_tfidf = LatentDirichletAllocation(n_components=5, random_state=42)
lda_tfidf.fit(tweet_word_matrix)
```

```python
pyLDAvis.sklearn.prepare(lda_tfidf, tweet_word_matrix, tfidf)
```
