## This next one includes only words in the nltk corpus, in an effort to only look at tweets in english

```python
X3 = data2['clean_english']

cv3 = CountVectorizer(stop_words = 'english')
X_cv3 = cv3.fit_transform(X3)

df3 = pd.DataFrame(X_cv3.toarray(), columns = cv3.get_feature_names())
df3.head()
```

```python
nmf = NMF(5)
nmf.fit(df3)
```

Tweet topic matrix

```python
W = nmf.transform(df3)
```

```python
W.shape
```

sort by topic 0 at first, and then look at the tweets in that topic

```python
pd.DataFrame(W).sort_values(0)
```

```python
top5 = [5275, 5263, 5349, 26, 3027]
for tweet in X3.iloc[top5]:
    print(tweet)
```

word topic matrix

```python
H = nmf.components_
H
```

```python
H.shape
```

```python
term_topic_matrix = pd.DataFrame(H, columns = cv3.get_feature_names()).T
term_topic_matrix
```

```python
term_topic_matrix.sort_values(0, ascending = False)
```

## Try the TFIDF vectorizer

#### Start by adding the word math to the stop words 

```python
from sklearn.feature_extraction import text

stop_words = list(text.ENGLISH_STOP_WORDS)
stop_words.append('math')
```

```python
type(stop_words)
```

```python
'math' in stop_words
```

```python
#just want to look at the clean english column
X3 = data2['clean_english']

#tokenize
tfid = TfidfVectorizer(stop_words = stop_words)
X_tfid = tfid.fit_transform(X3)

#put in a df
df4 = pd.DataFrame(X_tfid.toarray(), columns = tfid.get_feature_names())
df4.head()
```

```python
df4['math'] #check that 'math' was removed
```

```python
#fit an NMF to it
nmf2 = NMF(5, max_iter = 700)
nmf2.fit(df4)
```

```python
#W2 is the matrix of topics and the terms associated with each topic
W2 = nmf2.transform(df4)
```

```python
W2.shape
```

```python
#write a function to streamline this
def top_topic_tweets(df, topn, topic, X3 = X3):
    topic_1 = df.sort_values(topic, ascending = False)
    topic_topn = topic_1.iloc[:topn].index
    return X3.iloc[topic_topn]
```
