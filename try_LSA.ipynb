## Try LSA topic modeling

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer, ENGLISH_STOP_WORDS
from sklearn.decomposition import TruncatedSVD
from sklearn.decomposition import NMF
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Load in the data
df = pd.read_pickle('clean_tweets_no_duplicates.pkl')
df.head()
```

```python
corpus = df['clean_text3']
```

```python
tfidf = TfidfVectorizer(stop_words = ENGLISH_STOP_WORDS.union(['math', 'rt', 'https']))
tweet_word_matrix = tfidf.fit_transform(corpus)
vocab = tfidf.get_feature_names()
```

```python
tweet_word_matrix_lsa = pd.DataFrame(tweet_word_matrix.toarray(), columns = vocab)
```

```python
tweet_word_matrix_lsa.head()
```

```python
lsa = TruncatedSVD(n_components = 10)
tweet_topic_matrix = lsa.fit_transform(tweet_word_matrix)
lsa.explained_variance_ratio_
```

```python
lsa.explained_variance_ratio_.shape
```

```python
topic_word = pd.DataFrame(lsa.components_.round(3),
             index = ["component_"+str(i) for i in range(1,11)],
             columns = vocab)
topic_word
```

```python
def display_topics(model, feature_names, no_top_words, topic_names=None):
    for ix, topic in enumerate(model.components_):
        if not topic_names or not topic_names[ix]:
            print("\nTopic ", ix)
        else:
            print("\nTopic: '",topic_names[ix],"'")
        print(", ".join([feature_names[i]
                        for i in topic.argsort()[:-no_top_words - 1:-1]]))
```

## Check if the topics are interpretable

```python
display_topics(lsa, vocab, 20)
```

Looks like lots of repeated topics and not really as interpretable as NMF
