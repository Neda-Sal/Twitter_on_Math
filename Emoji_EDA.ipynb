# Goal1: Get a word bubble of the most common emojis

```python
import pandas as pd
import demoji
from wordcloud import WordCloud
from PIL import Image
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
tweets = pd.read_pickle('math_tweets_no_duplicates.pkl')
tweets.head()
```

```python
for tweet in tweets['text']:
    print(demoji.findall(tweet))
```

```python
from collections import defaultdict

for post in tweets['text']:
    if demoji.findall(post) != {}:
        print(demoji.findall(post))
```

## Use demoji to get the counts of all the emojis

```python
emoji_counts = defaultdict(int)
emoji_word_counts = defaultdict(int)

for post in tweets['text']:
    if demoji.findall(post) != {}:
        for key, value in demoji.findall(post).items():
            emoji_counts[key] += 1
            emoji_word_counts[value] += 1
emoji_counts
```

```python
emoji_word_counts
```

# Make a wordcloud
## Check if i can display the actual emojis

```python
wc = WordCloud(background_color="white", max_words=10,relative_scaling=0.5,normalize_plurals=False).generate_from_frequencies(emoji_counts)
plt.axis('off')
plt.imshow(wc)
```

## Nope, looks like it has to be just words

```python
wc = WordCloud(background_color="rgba(255, 255, 255, 0)", mode="RGBA", max_words=30,normalize_plurals=False).generate_from_frequencies(emoji_word_counts)
plt.figure(figsize = (10, 5))
plt.axis('off')
plt.imshow(wc)
plt.savefig('emoji_cloud.png');
```
