### Look at some basic PCA visuals

```python
# Imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.feature_extraction.text import TfidfVectorizer, ENGLISH_STOP_WORDS

%config InlineBackend.figure_format='retina'
```

```python
# Load in the data
df = pd.read_pickle('clean_tweets_no_duplicates.pkl')
df.head()
```

```python
corpus = df['clean_text3']
```

```python
stop_words = ENGLISH_STOP_WORDS.union(['math'])
tfidf = TfidfVectorizer(stop_words = stop_words)

tweet_word_matrix = tfidf.fit_transform(corpus)

vocab = tfidf.get_feature_names()
```

```python
pd.DataFrame(tweet_word_matrix.toarray(), columns = vocab)
```

```python
# Create a PCA instance: pca
pca = PCA(n_components=20)
principalComponents = pca.fit_transform(tweet_word_matrix)

# Plot the explained variances
features = range(pca.n_components_)
plt.bar(features, pca.explained_variance_ratio_, color='black')
plt.xlabel('PCA features')
plt.ylabel('variance %')
plt.xticks(features)

# Save components to a DataFrame
PCA_components = pd.DataFrame(principalComponents)
```

### Try to convert it to an array first

```python
# Create a PCA instance: pca
pca = PCA(n_components=20)
principalComponents = pca.fit_transform(tweet_word_matrix.toarray())

# Plot the explained variances
features = range(pca.n_components_)
plt.bar(features, pca.explained_variance_ratio_, color='black')
plt.xlabel('PCA features')
plt.ylabel('variance %')
plt.xticks(features)

# Save components to a DataFrame
PCA_components = pd.DataFrame(principalComponents)
```

```python
plt.scatter(PCA_components[0], PCA_components[1], alpha=.1, color='black')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
```

## How many clusters for k-means ?

```python
ks = range(1, 10)
inertias = []
for k in ks:
    # Create a KMeans instance with k clusters: model
    model = KMeans(n_clusters=k)
    
    # Fit model to samples
    model.fit(PCA_components.iloc[:,:3])
    
    # Append the inertia to the list of inertias
    inertias.append(model.inertia_)
    
plt.plot(ks, inertias, '-o', color='black')
plt.xlabel('number of clusters, k')
plt.ylabel('inertia')
plt.xticks(ks)
plt.show()
```

### Looks like there's an elbow point at 4 clusters
